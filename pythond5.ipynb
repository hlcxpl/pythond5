b. Genera una tabla que contenga la unión de las 3 tablas. hint: utiliza sufijos para para las
columnas que se llaman igual usando el parámetro suffixes de pd.merge().

hago un merge anidado de las tablas, primero hago un merge entre la tabla insidente y official con relacion en el numero de caso al igual que las tablas incidente y sujeto y despues le hago otro mergue para que las 3 tablas esten unidas por el numero de caso


```python
merged_df = pd.merge(
    pd.merge(
        incidents_df, officers_df, on="case_number", suffixes=("_incident", "_officer")
    ),
    subjects_df,
    on="case_number",
    suffixes=("_incident", "_subject"),
)
merged_df
#muestro el resultado del merge pero solo las titulos de las columnas y veo que al hacer el merge cambio los nombre de las columnas que eran repetidos con los subject y coloco 
#incident en vez de officer
merged_df.columns
```

a. Carga los datos y crea un DataFrame con ellos

```python
import pandas as pd
import numpy as np

# Cargar el archivo CSV y crear un DataFrame
df_jobs = pd.read_csv('Cleaned_DS_Jobs.csv')
df_jobs
```

b. Utiliza la siguiente lista de valores que serán considerados como nulos:

```python
# Eliminar filas con datos faltantes
df_jobs.dropna(inplace=True)
df_jobs
```

d. A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo
y Máximo. (hint: Utiliza el método apply sobre la columna.)


```python
df_jobs[['Salario Estimado Mínimo', 'Salario Estimado Máximo']] = df_jobs['Salary Estimate'].apply(
    lambda salary_string: pd.Series([int(s.replace('$', '').replace('K', '').strip()) for s in salary_string.split('-')])
    if '-' in salary_string
    else pd.Series([int(s.replace('$', '').replace('K', '').strip()), int(s.replace('$', '').replace('K', '').strip())])
)
df_jobs
```

e. Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint:
utilice reemplazo con diccionario usando el método replace sobre la columna.)
Size                          Descripción
10000+ employees          Mega Empresas
5001 to 10000 employees   Grandes Empresas
1001 to 5000 employees    Medianas Empresas
201 to 500 employees      Pequeñas Empresas
51 to 200 employees       Pequeñas Grandes Empresas
501 to 1000 employees     Microempresas
Unknown Empresas sin Información

```python
#Recodificación de la columna 'Size'
size_mapping = {
    '10000+ employees': 'Mega Empresas',
    '5001 to 10000 employees': 'Grandes Empresas',
    '1001 to 5000 employees': 'Medianas Empresas',
    '201 to 500 employees': 'Pequeñas Empresas',
    '51 to 200 employees': 'Pequeñas Grandes Empresas',
    '501 to 1000 employees': 'Microempresas',
    'Unknown': 'Empresas sin Información'
}
df_jobs['Size'] = df_jobs['Size'].replace(size_mapping)
df_jobs
```

f. Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y
la media del salario estimado máximo por tamaño de empresa. (hint: utiliza
pd.pivot_table para generar la vista adecuada con las columnas generadas.)

```python

# Generar la tabla pivote
pivot_table_size_salary = pd.pivot_table(df_jobs, index='Size',
                                         values=['Salario Estimado Mínimo', 'Salario Estimado Máximo'],
                                         aggfunc='mean')
pivot_table_size_salary
```
